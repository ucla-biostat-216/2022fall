# Plot: Four Fundamental Subspaces

```julia
using Luxor

@png begin
        # Box 1
        @layer begin
            θ = - 0.75(π / 4)
            c, s = cos(θ), sin(θ)
            transform([c, s, -s, c, 200, 0])
            box(Point(0, 0), Point(100, -150), :stroke)
            box(Point(0, 0), Point(-125, 75), :stroke)
        end
        # Box 2
        @layer begin
            θ = 1.25(π / 4)
            c, s = cos(θ), sin(θ)    
            transform([c, s, -s, c, -200, 0])
            box(Point(0, 0), Point(-100, -150), :stroke)
            box(Point(0, 0), Point(125, 75), :stroke)
        end
        # text top left
        @layer begin
            fontsize(16)
            text("C(A')", Point(-230, -160))
            text("dim r", Point(-230, -140))
            fontsize(12)
            text("row space", Point(-200, -90))
            text("all vectors A'y", Point(-210, -70))
        end
        # text left
        @layer begin
             settext("R<sup>n</sup>", Point(-230, 0), markup=true, valign="center")
        end
        # text bottom left
        @layer begin
            fontsize(16)
            text("N(A)", Point(-270, 110))
            text("dim n-r", Point(-280, 130))
            fontsize(12)
            text("null space", Point(-230, 70))
            text("Ax=0", Point(-220, 90))
        end
        # text top right
        @layer begin
            fontsize(16)
            text("C(A)", Point(240, -160))
            text("dim r", Point(240, -140))
            fontsize(12)
            text("column space", Point(160, -90))
            text("all vectors Ax", Point(170, -70))
        end
        # text right
        @layer begin
             settext("R<sup>m</sup>", Point(220, 0), markup=true, valign="center")
        end    
        # text bottom right
        @layer begin
            fontsize(16)
            text("N(A')", Point(200, 110))
            text("dim m-r", Point(190, 130))
            fontsize(12)
            text("left null space", Point(130, 70))
            text("A'y=0", Point(150, 90))
        end
        # title
        @layer begin
            fontsize(20)
            text("Four Fundamental Subspaces", Point(0, -230), halign=:center)
            settext("A is m-by-n", Point(0, 150), markup=true, halign="center")
        end
    end 600 600 "./four_fundamental_subspaces.png"
```

```julia
using Luxor

@png begin
    fontsize(15)
    # box
    rect(Point(-2cm, 2cm), 10cm, -10cm, :stroke)
    line(Point(-1cm, 2cm), Point(4cm, -8cm), :stroke)
    line(Point(3cm, -6cm), Point(3cm, -2cm), :stroke)
    line(Point(1cm, -2cm), Point(3cm, -2cm), :stroke)
    line(Point(1.4cm, -2.8cm), Point(3cm, -2cm), :stroke)
    label("(0, 0)", :E, O)
    label("(3, 2)", :E, Point(3cm, -2cm))
    label("(3, 6)", :E, Point(3cm, -6cm))
    label("(1, 2)", :W, Point(1cm, -2cm))
    label("(1.4, 2.8)", :W, Point(1.4cm, -2.8cm))
    label("C(X)", :W, Point(2cm, -4cm))
    for xcoord in -2:1:8
        label("$xcoord", :S, Point(xcoord * 1cm, 2cm))
        label("$xcoord", :W, Point(-2cm, -xcoord * 1cm))
    end
#     rulers()
    end 500 500 "./three_projs.png"
```
